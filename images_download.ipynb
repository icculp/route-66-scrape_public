```python
import requests
from pandas import read_csv
from os import mkdir

# Make images directory
try:
  mkdir('images')
except:
  pass

# Open places_final as a dataframe
places_df = read_csv('places_final.txt', sep='|')


failed = []
missing = []
bad_image_url = 'https://www.historic66.com/templates/data/images/sharepoi.svg'
# Loop through dataframe
for i, row in places_df.iterrows():
  # Get image url
  image_url = row['Image']
  if image_url == bad_image_url:
    missing.append(row['RoadTripStopName'])
  
  # Download image
  response = requests.get(image_url, stream=True)

  # Save image to images folder
  if response.status_code == 200:
    image_name = 'images/{}.jpg'.format(row['PlaceID'])
    with open(image_name, 'wb') as image_file:
        for chunk in response:
            image_file.write(chunk)
    print('Image for {} complete'.format(row['RoadTripStopName']))
  else:
    print('Image not found for {}'.format(row['RoadTripStopName']))
    failed.append(row['RoadTripStopName'])

# Print anything that didn't work 
with open('places_missing_images', 'w') as fail_file:
  print(missing)
  print(failed)
  fail_file.write('failed to download:\n')
  fail_file.write(str(failed))
  fail_file.write('image unavailable:\n')
  fail_file.write(str(missing))



```
